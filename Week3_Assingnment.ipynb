**In this notebook i will explore, segment, and cluster the neighborhoods in the city of Toronto**

```python
#export data and ignore 'not assigned' data
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
dropi = df[((df['Borough'] == 'Not assigned'))].index
df.drop(dropi, inplace = True)
df.reset_index(drop = True, inplace = True)

df.head()
```

# Combine first dataset with longitude and latitude data

```python
df1 = df.sort_values('Postal Code')
df1 = df1.reset_index(drop = True)

df3 = pd.concat([df1, df2[['Latitude', 'Longitude']]], axis = 1)
df3.head()
```

# Visualize Toronto map for every postal code

```python
latitude = 56.1304
longitude = -106.3468
canada_map = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df3['Latitude'], df3['Longitude'], df3['Borough'], df3['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(canada_map)  
    
canada_map
```

```python
toronto_data = df3[df3['Borough'].str.contains('Toronto')].reset_index(drop=True)
toronto_data.head()
```

# Visualize Toronto map with 'Toronto' word in borough

```python
latitude = 43.6532
longitude = -79.3832
toronto_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# add markers to map
for lat, lng, borough, neighborhood in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Borough'], toronto_data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_map)  
    
toronto_map
```
