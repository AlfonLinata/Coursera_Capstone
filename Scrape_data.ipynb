```python
details = []
```

```python
condition = True
while condition:

    product = webD.find_elements_by_class_name('header-container') #find all product(house) element
    webD.execute_script('window.scrollBy(0,5100)','') #scroll 5100 pixel, I find this make our program mmore undetecable
    
    #loop through every item to get all the details
    for i, val in enumerate(product):
        p2 = val.find_element_by_tag_name('a')
        
        #try except program used to avoid program to stop if there is no data input
        try:
            name = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/div[1]/div/h3/a'.format(str(i+1))).text
        except:
            name = None
        try:
            address = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/div[1]/div/p'.format(str(i+1))).text
        except:
            address = None
        try:
            bedroom = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/ul[2]/li[1]/span[1]'.format(str(i+1))).text
        except:
            bedroom = 0
        try:
            bathroom = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/ul[2]/li[1]/span[2]'.format(str(i+1))).text
        except:
            bathroom = 0
        try:
            area = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/ul[2]/li[2]'.format(str(i+1))).text
        except:
            area = None
        try:
            price_m2 = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/ul[2]/li[3]'.format(str(i+1))).text
        except:
            price_m2 = None
        try:    
            price = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[1]/div[2]/ul[1]'.format(str(i+1))).text
        except:
            price = None
        try:
            description = webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[2]/div[{}]/div[2]/div/div/div[1]/div[1]/div[2]'.format(str(i+1))).text
        except:
            description = None
            
        # Make dictionary of the house detail and then store it
        temp = {
        "name": name,
        "address": address,
        "bedroom": bedroom,
        "bathroom": bathroom,
        "area": area,
        "price/m2": price_m2,
        "price": price,
        "description":description,
        "link": p2.get_property('href')
        }
        details.append(temp)
        
    #click the next page button, if there is no next page button stop the loop
    try:
        #webD.find_element_by_xpath('//*[@id="wrapper-inner"]/section[1]/div[2]/div/div[2]/div[2]/section/div[3]/ul/li[{}]/a'.format(str(j+6))).click()
        webD.find_element_by_class_name('pagination-next').find_element_by_tag_name('a').click()
    except:
        condition = False
        
    print(len(details))
# this program still will be interupted every around 68 pages to make sure you are not bot
```

```python
df = pd.DataFrame(details)
df.to_csv("Bandung_House_Price.csv")
```
